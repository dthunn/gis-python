```python
from shapely.geometry import Point, LineString, Polygon, MultiPoint, MultiLineString, MultiPolygon
```

```python
point = Point(2, 3)
point
```

```python
line = LineString([(0,0), (1,2), (2,3)])
line
```

```python
polygon1 = Polygon([(0,0), (4,0), (4,3), (0,3), (0,0)])
polygon1
```

```python
polygon2 = Polygon([(0,0), (4,0), (4,3), (0,3), (0,0), (-1,-1)])
polygon2
```

```python
#polygons with holes
hole = [(1,1), (2,1), (2,2), (1,2), (1,1)]

polygon_with_hole = Polygon(shell=[(0,0), (4,0), (4,3), (0,3), (0,0)], holes=[hole])
polygon_with_hole
```

```python
polygon_with_hole.is_valid
```

```python
multipoint = MultiPoint([(1,1), (2,2), (3,3)])
multipoint
```

```python
point1 = Point(1,1)
point2 = Point(3,2)
point3 = Point(4,5)
point4 = Point(10,10)
multipoint = MultiPoint([point1, point2, point3, point4])
multipoint
```

```python
line1 = LineString([(0,0), (1,2), (2,3)])
line2 = LineString([(3,3), (5,3)])

multiline = MultiLineString([line1, line2])
multiline
```

```python
polygon1 = Polygon([(0,0), (4,0), (4,3), (0,3), (0,0)])
polygon2 = Polygon([(0,0), (4,2), (5,3), (1,3), (0,0)])

multipolygon = MultiPolygon([polygon1, polygon2])
multipolygon
```

```python
polygon1.area
```

```python
print(line1.length)
print(line2.length)
print(multiline.length)
print(polygon1.length)
print(multipolygon.length)
```

```python
polygon1.centroid
```

```python
for polygon in multipolygon.geoms:
    print(list(polygon.exterior.coords))
```
