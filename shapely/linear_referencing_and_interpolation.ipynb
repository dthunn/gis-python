```python
distances = [1, 2, 3, 5, 7]
```

```python
interpolated_points = []

for distance in distances:
    interpolated_points.append(line2.interpolate(distance))
```

```python
interpolated_points
```

```python
multi_point = MultiPoint(interpolated_points)
```

```python
multi_point
```

```python
plot_geometries(line2, multi_point)
```
